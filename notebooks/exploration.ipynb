# Merchant Retention Exploration

Lightweight notebook for inspecting the generated merchant portfolio and saved scoring outputs.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

PROJECT_ROOT = Path.cwd() if (Path.cwd() / 'src').exists() else Path.cwd().resolve().parent
raw_path = PROJECT_ROOT / 'data' / 'raw' / 'merchants.csv'
output_path = PROJECT_ROOT / 'outputs' / 'tables' / 'merchant_retention_actions.csv'

raw_df = pd.read_csv(raw_path)
scored_df = pd.read_csv(output_path)

raw_df.head()
```

```python
scored_df[['recommended_action', 'expected_retention_value']].groupby('recommended_action').agg(['count', 'mean', 'sum'])
```

```python
plt.figure(figsize=(8, 5))
plt.hist(scored_df['churn_probability'], bins=25, color='#1f77b4', edgecolor='white')
plt.title('Churn Probability Distribution')
plt.xlabel('Churn Probability')
plt.ylabel('Merchant Count')
plt.show()
```
